# Quantum Progam Using Gate Matrices

## Gates Implemented:

- Identity Gate: qp.i(qubit)

- Hadamard Gate: qp.h(qubit)

- Phase Gate: qp.phase(theta, qubit)

- Pauli X Gate: qp.x(qubit)
- Pauli Y Gate: qp.y(qubit)
- Pauli Z Gate: qp.z(qubit)

- X Rotation Gate: qp.rot_x(theta, qubit)
- Y Rotation Gate: qp.rot_y(theta, qubit)
- Z Rotation Gate: qp.rot_z(theta, qubit)

- CNOT Gate: qp.cnot(control_qubit, target_quibit)

- Toffoli Gate: qp.t(control_qubit1, control_qubit2, target_qubit)

```python
from Matrix_Program import *
```

```python
# Start program with 2 qubits, all set to |0>
qp = qprogram(2)

# Apply hadamard gate to qubit 0
qp.h(0)

# Run program
qp.run(show_states=False, plot=True)
```

```python
# Start program with 3 qubits, all set to |0>
qp = qprogram(3)

# Apply hadamard gate to qubit 0
qp.h(0)

# Apply hadamard gate to qubit 2
qp.h(2)

# Run program
qp.run(show_states=False, plot=True)
```

```python
# Start program with 3 qubits, qubit 0 |0>, qubit 1 |1>, qubit 2 |1>
qp = qprogram([0, 1, 1])

# Apply hadamard gate to qubit 0
qp.h(0)

qp.run(show_states=False, plot=True)
```

```python
# Start program with 3 qubits, qubit 0 |0>, qubit 1 |1>, qubit 2 |1>
qp = qprogram([0, 1, 1])

# Apply several gates
qp.h(0)
qp.h(2)
qp.x(0)
qp.cnot(0, 1)

# Run program with states shown
qp.run(show_states=True, plot=True)
```
